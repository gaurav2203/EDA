```python
df_country= pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df= pd.merge(df, df_country, on= 'Country Code', how='left')
```

```python
final_df.head(2)
```

```python
##To Check Data Types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names= final_df.Country.value_counts().index
```

```python
country_val= final_df.Country.value_counts().values
```

```python
## Pie Chart
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
```

## Observation
1. Zomato max records are from India then USA and UK

```python
final_df.columns
```

```python
ratings= final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observations
1. Whenever the rating is between 4.5 to 3.9 --> Excellent
2. 4.0 to 3.4 --> Very Good
3. 3.5 to 3.9 --> Good
4. 3.0 to 3.4 --> Average
5. 2.5 to 2.9 --> Average
6. 2.0 to 2.4 --> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']= (12, 6)
sns.barplot(x='Aggregate rating', y= 'Rating Count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y= 'Rating Count', data=ratings, hue='Rating color', palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```

## Observation
1. Not rated count is very high.
2. Max ratings are between 2.5 to 3.4

```python
## Count Plot
sns.countplot(x= 'Rating color', data=ratings, palette= ['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```
