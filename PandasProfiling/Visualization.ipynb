```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x= np.arange(0, 10)
y= np.arange(11, 21)

a= np.arange(40, 50)
b= np.arange(50, 60)
```

```python
# Plot Scatter
plt.scatter(x, y, c='r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Scatter Plot')
```

```python
# Plot
y= x*x
plt.plot(x, y, 'ro--', linewidth=2, markersize= 12)
```

```python
# Subplots
plt.subplot(2,2,1)
plt.plot(x, y, 'r')
plt.subplot(2,2,2)
plt.plot(x, y, 'g')
plt.subplot(2,2,3)
plt.plot(x, y, 'b')
```

```python
x= np.arange(0, 5* np.pi, 0.1)
y= np.sin(x)
plt.subplot(2, 1, 1)
plt.title('Sine Curve')
plt.plot(x, y, 'r')

plt.subplot(2, 1, 2)
y= np.cos(x)
plt.title('Cosine Curve')
plt.plot(x, y, 'g')
```

```python
#Bar Graph
x= [2, 8, 10]
y= [11, 16, 9]

x2= [3, 9, 11]
y2= [6, 15, 7]

plt.bar(x, y)
plt.bar(x2, y2)
plt.title('Bar Graph')
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

```python
#Histograms
x= np.array([22, 87, 5, 43, 56, 73, 55, 54, 11, 20, 51, 5, 79, 31, 27])
plt.hist(x)
plt.title('Histrgram')
plt.show()
```

```python
#Box Plots
data= [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, vert= True, patch_artist= True)
```

```python
#JoinPlot
sns.jointplot('tip', 'total_bill', data= df, kind= 'hex')
```
